# Take-home problems for this week:

1- Filter the data to include only weekdays (Monday to Friday) and plot a line graph showing the pedestrian counts for each day of the week.

```python
# Convert the hour_beginning column to datetime
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'], errors='coerce')

# Extract time components
df['hour'] = df['hour_beginning'].dt.hour
df['month'] = df['hour_beginning'].dt.month
df['date'] = df['hour_beginning'].dt.date
df['day_name'] = df['hour_beginning'].dt.day_name()

# Quick check
df[['hour_beginning', 'hour', 'month', 'date', 'day_name']].head()
```

```python
# Keep only Monday–Friday
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
df_weekdays = df[df['day_name'].isin(weekdays)]

# Verify
df_weekdays['day_name'].value_counts()
```

```python
daily_sum = df_weekdays.groupby('date')['Pedestrians'].sum().reset_index()

daily_sum['day_name'] = pd.to_datetime(daily_sum['date']).dt.day_name()

weekday_avg_per_day = daily_sum.groupby('day_name')['Pedestrians'].mean().reindex(weekdays)

weekday_avg_per_day
```

```python
plt.figure(figsize=(7,4))
plt.plot(weekday_avg_per_day.index, weekday_avg_per_day.values, color='navy', marker='o', linewidth=2)
plt.title('Average Daily Pedestrian Counts by Weekday (Brooklyn Bridge)', fontsize=13)
plt.xlabel('Day of Week', fontsize=11)
plt.ylabel('Average Daily Pedestrians', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

2- Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year. Sort the pedestrian count data by weather summary to identify any correlations( with a correlation matrix) between weather patterns and pedestrian counts for the selected year.

```python
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'], errors='coerce')

df_2019 = df[df['hour_beginning'].dt.year == 2019]

df_2019['hour_beginning'].min(), df_2019['hour_beginning'].max()
```

```python
weather_pedestrians = df_2019.groupby('weather_summary')['Pedestrians'].sum().sort_values(ascending=False)

weather_pedestrians
```

3- Implement a custom function to categorize time of day into morning, afternoon, evening, and night, and create a new column in the DataFrame to store these categories. Use this new column to analyze pedestrian activity patterns throughout the day.

```python
def categorize_time(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'
```

```python
df.columns
```

```python
df['time_of_day'] = df['hour'].apply(categorize_time)
```

```python
df[['hour', 'time_of_day']].head()
```

```python
pedestrians_by_time = df.groupby('time_of_day')['Pedestrians'].sum().reindex(['Morning','Afternoon','Evening','Night'])
pedestrians_by_time
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x=pedestrians_by_time.index, y=pedestrians_by_time.values, palette='viridis')
plt.title('Pedestrian Counts by Time of Day (Brooklyn Bridge)', fontsize=14)
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Total Pedestrians', fontsize=12)
ax = plt.gca()
ax.get_yaxis().get_major_formatter().set_scientific(False)
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%d'))
plt.tight_layout()
plt.show()
```
